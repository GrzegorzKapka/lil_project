```python
# I want to share with you one of my little project that I made while I was learning to use pandas.

# My approach to filter the data (I am mainly focusing on fee, club_name and club_involved_name columns) (Ignoring "fee_cleaned" column just to make it a bit more challanging)
# 1) To save only rows with paid transfers for future plots, dropping unknown data and loan transfers.
#    Change the fee data from string to int.
# 2) To create ("to club", "from club") in place of ("club_name", "club_involved_name", "transfer_movement")
#    ("transfer_movement" informs whether the direction of a transfer was from "club_name" or "club_involved_name")
import pandas as pd
import requests
```

```python
# download the data from github.com
r = requests.get("https://raw.githubusercontent.com/ewenme/transfers/master/data/premier-league.csv")
```

```python
file = open("data.csv", "w")
file.write(r.text)
file.close()
```

```python
# read the data
df = pd.read_csv("data.csv")
df.drop(columns = ["fee_cleaned"], inplace = True)
```

```python
df.rename(columns={"club_involved_name": "to club", "club_name":"from club"}, inplace=True)
```

```python
# I found out that I could not do much with missing information so I decided to delete all rows with them.
filt2 = (df["fee"].apply(type) != str)
```

```python
filt3 = (df[~filt2]["fee"].apply(lambda x: x[0]) == '€')
```

```python
# Filtering out all the rows that were not paid transfers.
df = df[~filt2][filt3]
```

```python
currency = df["fee"].apply(lambda x: x[0])
```

```python
# while working only with this data this is not necessary, it could be useful in the future updates of this small project.
df["currency"] = currency
```

```python
# removing prefix with the information of a currency
df["fee"] = df["fee"].apply(lambda x: x[1:])
```

```python
# creating a function that will transition strings to integers in "fee" column.
def fee_to_int(fee):
    if fee.find("m") != -1:
        return int(fee.replace("m","").replace(".","")) * 10000
    elif fee.find("Th") != -1:
        return int(fee.replace("Th","").replace(".","")) * 1000
    elif fee.find("k") != -1:
        return int(fee.replace("k","").replace(".","")) * 1000
    else:
        return fee
```

```python
df["fee"] = df["fee"].apply(fee_to_int)
```

```python
# Transfer fees in descending order.
df.sort_values(by = "fee", ascending = False)
```

```python
# Now I will take care of "transfer_movement" column.
df["transfer_movement"].unique()
```

```python
# I'm simply changing values from one to another under the condition whether transfer is "in" or "out"
df_out = df[df["transfer_movement"] == "out"]
```

```python
df_else = df[df["transfer_movement"] != "out"]
```

```python
new_to_club = df_out["from club"]
```

```python
df_out["new_to_club"] = new_to_club
```

```python
# Dropping unnecessary column.
df_out.drop(columns = ["from club"], inplace = True)
```

```python
df_out.rename(columns={"to club": "from club", "new_to_club":"to club"}, inplace=True)
```

```python
df = df_else.append(df_out)
```

```python
df.sort_index(inplace = True)
```

```python
df.drop(columns = ["transfer_movement"], inplace = True)
```

```python
df = df[['player_name', 'from club', 'to club', 'age', 'position', 'currency', 'fee',
         'season', 'year', 'transfer_period', 'league_name', 'country']]
```
